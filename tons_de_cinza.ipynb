```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
foto = cv.imread('/content/windows.png') #imagem original

cv2_imshow(foto)
```

```python
foto_cinza = cv.cvtColor(foto, cv.COLOR_BGR2GRAY) #funo que transforma pra cinza
cv2_imshow(foto_cinza)
```

Agora, testarei a implementao manual convertendo os valores das clulas das matrizes

```python
dimensions = foto.shape
print(dimensions) #dimenses

foto_cinza_manual = foto.copy() #levei pau pq quando iguala altera nos 2
#pode iniciar como um array de zeros com np.zeros((dimensions[0], dimensions[1], dimensions[2]), np.uint8)

for i in range(dimensions[0]): #altura
  for j in range(dimensions[1]): #largura
    c = ((59 * foto[i,j,1]) + (29 * foto[i,j,0]) + (11 * foto[i,j,2])) / 100 #c = 0,59 * G + 0,29 * R + 0,11 * B
    for k in range(dimensions[2]): #canais (RGB)
      foto_cinza_manual[i,j,k] = c #cada valor da matriz RGB naquele pixel vira c

cv2_imshow(foto)
cv2_imshow(foto_cinza_manual)
```
