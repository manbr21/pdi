```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import math as mth
from google.colab.patches import cv2_imshow
```

```python
foto = cv.imread('/content/windows.png')
cv2_imshow(foto)
```

Tentativa de transformar para CMYK

```python
size = foto.shape

foto_cmyk = np.zeros((size[0],size[1],4)) #mesmas dimenses, porm com 1 matriz a mais para o preto

print(size)

for i in range(size[0]):
  for j in range(size[1]):

    R = foto[i,j,0] / 255
    G = foto[i,j,1] / 255
    B = foto[i,j,2] / 255

    K = 1 - max(R,G,B)
    if(K != 1):
      C = (1 - R - K) / (1 - K)
      M = (1 - G - K) / (1 - K)
      Y = (1 - B - K) / (1 - K)
    else:
      C = (1 - R - K)
      M = (1 - G - K)
      Y = (1 - B - K)

    foto_cmyk[i,j,0] = C * 255
    foto_cmyk[i,j,1] = M * 255
    foto_cmyk[i,j,2] = Y * 255
    foto_cmyk[i,j,3] = K * 255


cv2_imshow(foto)
cv2_imshow(foto_cmyk)
```
